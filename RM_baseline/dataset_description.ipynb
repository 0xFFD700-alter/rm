## 数据集概览

训练数据集包含90个数据文件，每个文件为一个用户所享有的数据集，用户之间不能共享数据，其目录如下图所示：

每个文件有多个样本组成，每个样本的数据存储格式为<位置，信号强度>。其中位置表示用户当前位置，是维度为(1,2)的实数，信号强度为该用户与4个基站之间的信号强度，是维度为(1,4)的实数。当基站与用户之间信号不可达时，信号强度设置为0。读取单个文件的示例代码如下：

```python
import numpy as np

data_user_01=np.reshape(np.load('./data/train/user_01.npy'),(-1,6))
print(f'data_user_01.shape: {data_user_01.shape}')
print(f'data_user_01: {data_user_01}')
```
